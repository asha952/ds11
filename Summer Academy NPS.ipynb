```python
import pandas as pd

df = pd.read_csv('SA Feedback Surveys_FINAL/2017/Student Feedback Surveys-Superview.csv')
```

```python
df.head(10)
```

# Clean up data
### Convert Ratings to numbers

```python
# Drop rows that contain #ERROR! in Rating (Num) column
df = df[~df['Rating (Num)'].isin(['#ERROR!'])]

# Convert elements in Rating (Num) to int
df['Rating (Num)'] = pd.to_numeric(df['Rating (Num)'])
```

### Add a column that represents the NPS class

```python
df['NPS Sentiment'] = df.apply(lambda row: ('Promoter' if row['Rating (Num)'] > 8 else 
                                            'Passive' if row['Rating (Num)']  > 6 else 'Detractor'), axis=1)
```

### Convert elements in week column to integers

```python
# Strip string, then split string, then convert the last element to integer
df['Week'] = df.apply(lambda row: int(row['Week'].strip().split(' ')[-1]), axis=1)
```

### Add a column that represents Schedule Pacing as a value

-2 | -1 | 0 | 1 | 2
--- | --- | --- | --- | ---
Way too slow | A little too slow | Just right | A little too fast | Way too fast


```python
pacing_map = {
    'Way too slow': -2, 
    'A little too slow': -1,
    'Just right': 0,
    'A little too fast': 1,
    'Way too fast': 2
}
df['Pacing Value'] = df.apply(lambda row: pacing_map.get(row['Schedule Pacing']), axis=1)
```

# Calculate the NPS

```python
promoter_count = len(df[df['NPS Sentiment'] == 'Promoter'])
passive_count = len(df[df['NPS Sentiment'] == 'Passive'])
detractor_count = len(df[df['NPS Sentiment'] == 'Detractor'])

print(f'Difference between the number of promoters and detractors: {promoter_count - detractor_count}')
```

```python
NPS = (promoter_count - detractor_count) / (promoter_count + passive_count + detractor_count)
print(f'Net Promoter Score: {NPS}')
```

### NPS by Location

```python
def cat_nps(column_name, category_name):
    ## Calculates the NPS for category in a column
    promoter_count = len(df[(df[column_name] == category_name) & (df['NPS Sentiment'] == 'Promoter')])
    passive_count = len(df[(df[column_name] == category_name) & (df['NPS Sentiment'] == 'Passive')])
    detractor_count = len(df[(df[column_name] == category_name) & (df['NPS Sentiment'] == 'Detractor')])
    
    return (promoter_count - detractor_count) / (promoter_count + passive_count + detractor_count)
```

```python
locations = df['Location'].dropna().unique()
locations_nps = map(lambda location: column_nps('Location', location), locations)
df_loc_nps = pd.DataFrame(zip(*(locations, locations_nps)), columns=('Location', 'NPS'))

df_loc_nps
```

```python
sns.barplot(x='NPS', y='Location', data=df_loc_nps)
```

### NPS by Track

```python
tracks = df['Track'].dropna().unique()
tracks_nps = map(lambda track: column_nps('Track', track), tracks)
df_track_nps = pd.DataFrame(zip(*(tracks, tracks_nps)), columns=('Track', 'NPS'))
```

### NPS by Week

```python
weeks = df['Week'].dropna().unique()
weeks_nps = map(lambda week: cat_nps('Week', week), weeks)
df_week_nps = pd.DataFrame(zip(*(weeks, weeks_nps)), columns=('Week', 'NPS'))
df_week_nps
```

```python
sns.lmplot(x='NPS', y='Week', data=df_week_nps)
```
